```python
def movie_classifier():
    
    n = 0
    ratings['class'] = 'undefined'
    for rating in ratings['rating']:
        if rating <= 2.0:
            ratings['class'][n] = 'низкий рейтинг'
        if 2.0 < rating <= 4.0:
            ratings['class'][n] = 'средний рейтинг'
        if rating >= 4.5:
            ratings['class'][n] = 'высокий рейтинг'
        n += 1
        if n == 500:
            break     
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings.head()
```

```python
movie_classifier()
```

```python
user_data = ratings.groupby('userId').count()
user_data.head()
```

```python
user_data = user_data[ user_data['class'] >= 100 ]
user_data.head()
```

```python
timestamp_data = ratings.groupby('userId').agg(['min', 'max'])['timestamp'].head()
```

```python
years=list(range(1950,2011))
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
def production_year(line):
    n=0
    movies['year'] = '1900'
    for line in movies['title']:
        for year in years:
            if str(year) in line:
                movies['year'][n] = year
                break
        n += 1
```

```python
def production_year(title):
    for year in years:
        if str(year) in title:
            return year
            return '1900'
```

```python
movies['year']=movies.title.apply(production_year)
```

```python
movies_and_ratings = movies.merge(ratings, left_index=True, right_index=True)
movies_and_ratings.head()
```

```python
movies_and_ratings.groupby('year').rating.mean().sort_values(ascending=False)
```
